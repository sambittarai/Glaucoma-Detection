```python
import os

def get_path_list(root_path,img_path,label_path):
    tmp_list = [img_path,label_path]
    res = []
    for i in range(len(tmp_list)):
        data_path = os.path.join(root_path,tmp_list[i])
        filename_list = os.listdir(data_path)
        filename_list.sort()
        res.append([os.path.join(data_path,j) for j in filename_list])
    return res

def write_path_list(name_list, save_path, file_name):
    f = open(os.path.join(save_path, file_name), 'w')
    for i in range(len(name_list[0])):
        f.write(str(name_list[0][i]) + " " + str(name_list[1][i]) + '\n')
    f.close()
```

```python
if __name__ == "__main__":
    #------------Path of the dataset --------------------------------
    data_root_path = '/content/drive/MyDrive/Glaucoma_Detection/Optic_Disk_patches'
    if not os.path.exists(data_root_path):
        raise ValueError("data path doesnot exist")
    #train
    img_train = "Training/Image_Patches"
    gt_train = "Training/Optic_Disc_GT_Patches"

    save_path = "/content/drive/MyDrive/Glaucoma_Detection/Dataset_Preparation/data_path_list/IDRiD"
    if not os.path.exists(save_path):
        os.mkdir(save_path)
    #-----------------------------------------------------------------
    train_list = get_path_list(data_root_path,img_train,gt_train)
    print('Number of train imgs:',len(train_list[0]))
    write_path_list(train_list, save_path, 'train.txt')

    print("Done")
```

# Check if the path exists or not

```python
train_path = "/content/drive/MyDrive/Glaucoma_Detection/Dataset_Preparation/data_path_list/IDRiD/train.txt"

img_list, gt_list = [], []
with open(train_path, 'r') as file_to_read:
  while True:
    lines = file_to_read.readline().strip()
    if not lines:
      break
    img,gt = lines.split(' ')
    img_list.append(img)
    gt_list.append(gt)

len(img_list), len(gt_list)
```

```python
# Path Exists or not

img = 0
gt = 0
for i in range(len(img_list)):
  if os.path.exists(img_list[i]):
    img += 1
  if os.path.exists(gt_list[i]):
    gt += 1

img, gt
```
